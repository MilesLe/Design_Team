```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from os import listdir
from os.path import isfile, join
```

```python
# Data extraction functions
file_name = "1.2_2_trial5_09102022.csv"
file_path = "./data/" + file_name

df = pd.read_csv(file_path)
#display(df)

force_cols = [col for col in df.columns if 'Force' in col]
```

```python
# Analysis functions

trials = []
max_forces = []
count = 0
for col in force_cols:
    count += 1
    trials.append(count)
    max_forces.append(max(df[col].values))

d = {'trial': trials, 'max_force': max_forces}
my_df = pd.DataFrame(data=d)
#display(my_df)
```

```python
# Apply analysis functions

sns.set_style("whitegrid")
sns.set(rc={'figure.figsize':(15,10)})
# plot the mean line
ax = sns.boxplot(meanline=True, y=my_df["max_force"])
#ax.set(ylim=(0, 10))
plt.show()
```
